```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

Data Collection and Data Processing

```python
landslide_data = pd.read_csv('../data_set/landslide_dataset.csv', header=0)
```

```python
landslide_data.head()
```

```python
landslide_data.shape
```

```python
landslide_data.describe()
```

```python
landslide_data['Landslide'].value_counts()
```

1 -> Landslide  
0 -> No landslide

```python
landslide_data.groupby('Landslide').mean()
```

```python
# seprating data and Labels
X = landslide_data.drop(columns='Landslide', axis=1)
Y = landslide_data['Landslide']
```

```python
print(X)
print(Y)
```

Training and Test data  

X_train = training data  
X_test = testing data  
Y_train = label of training data  
Y_test = label of testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=1)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

Model Training --> Logistic Regression

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data: ", training_data_accuracy)
```

```python
# accuracy on testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test data : ", test_data_accuracy)
```
